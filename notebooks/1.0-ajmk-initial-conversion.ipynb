```python
import pandas as pd
import numpy as np
```

```python
columns = ["index", "Variable","Dim1","Dim2","Dim3","Dim4","Dim5","Dim6","Dim7","Dim8","Dim9","Dim10","ResultValue"]
results_dir = "/Users/alexanderkell/Documents/SGI/Projects/17-ClicSAND/data/raw/osemosys_output/HO9_results.txt"
osemosys_output = pd.read_csv(results_dir, names=columns, sep="\(|,|\)|[ \t]{1,}", engine="python")
osemosys_output.head(1000)
```

```python
def check_for_all_zeros(df):
    df.loc[:, (df == 0).all(axis=0)] = np.nan
#     df['ResultValue'] = df.apply(pd.Series.last_valid_index, axis=0).Variable
    return df

osemosys_clean = osemosys_output.groupby("Variable").apply(lambda x: check_for_all_zeros(x))
osemosys_clean
```

```python
osemosys_clean["ResultValue"] = osemosys_clean.ffill(axis=1).iloc[:, -1]
osemosys_clean.head(1000)
```

```python
def replace_result_value(df):
    df[df.loc[:, ~df.isnull().all()].iloc[:, -2].name] = np.nan
    return df

osemosys_cleaned = osemosys_clean.groupby("Variable").apply(lambda x: replace_result_value(x))
osemosys_cleaned.head(1000)
```

```python
osemosys_cleaned = osemosys_cleaned.drop("index", axis=1)
osemosys_cleaned
```

```python
osemosys_cleaned.to_csv("/Users/alexanderkell/Documents/SGI/Projects/17-ClicSAND/data/processed/output.csv", index=False)
osemosys_cleaned.head(1000)
```
